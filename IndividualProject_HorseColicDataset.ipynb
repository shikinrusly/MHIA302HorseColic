```python
#k. Read and evaluate the following code in Figure 1. What method is the code using to
#reduce the missing value from 1605 to 0? Justify your answer.
# the Part 2 (L)"Fill out the inline comments marked with hashtags ‘#’ in lines 8, 11, 17, 20, 22 and 26. 
#Include the missing codes in lines 5 and 6." is using the same dataset as (a) to (k). 
from numpy import isnan
from pandas import read_csv
from sklearn.impute import SimpleImputer

#load dataset
#replace the '?' with NaN
import pandas as pd
filename = '/content/drive/My Drive/Colab Notebooks/MHIA302/horsecolicdata.txt'
df = pd.read_csv(filename,  sep="\t", na_values = '?',  header=None)
df
df6 = df.replace(['?.1', '?.2', '?.3','?.4', '?.5', '?.6','?.7','?.8','?.9'], np.NaN)
print(df6)
Final_df = df6.astype('float')
Final_df
```

```python
#To replace the missing values of each column using mean of the data along each column

imputer = SimpleImputer(strategy='mean')
```

```python
#to transform the data into a new dataset replacing the NaN value to the mean value of each of the column
imputer.fit(X)
Xtrans = imputer.transform(X)

```
